# Ames Housing EDA Steps
# Part 1: Import Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style for consistency
sns.set(style="whitegrid")

# Load the Ames Housing dataset
# Update the path if needed
DATA_PATH = '/kaggle/input/ames-housing-dataset/AmesHousing.csv'
df = pd.read_csv(DATA_PATH)

# Display the shape and first few rows
print("Shape of dataset:", df.shape)
display(df.head()) 
```

# Part 2: Data Overview & Summary Statistics

```python

# Display the shape and first few rows
print("Shape of dataset:", df.shape)
display(df.head())

# Part 2: Data Overview & Summary Statistics

# Show info about the DataFrame
print("\nDataFrame Info:")
df.info()

# Show summary statistics for numeric columns
print("\nSummary Statistics:")
display(df.describe())

# Check for missing values in each column
print("\nMissing Values per Column:")
missing = df.isnull().sum()
missing = missing[missing > 0]
print(missing) 
```

# Part 3: Univariate Analysis

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")


df.replace([np.inf, -np.inf], np.nan, inplace=True)

# List of key variables to analyze
key_vars = ['SalePrice', 'Lot Area', 'Year Built', 'Gr Liv Area']

for var in key_vars:
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df[var].dropna(), kde=True)
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[var].dropna(), orient='h')
    plt.title(f'Boxplot of {var}')
    plt.xlabel(var)

    plt.tight_layout()
    plt.show()

    # Print summary statistics for each variable
    print(f'\nSummary statistics for {var}:')
    print(df[var].describe()) 
```

# Part 4: Bivariate Analysis

```python


# Numeric variables vs. SalePrice
num_vars = ['Lot Area', 'Gr Liv Area', 'Year Built']
for var in num_vars:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=df[var], y=df['SalePrice'])
    plt.title(f'{var} vs. SalePrice')
    plt.xlabel(var)
    plt.ylabel('SalePrice')
    plt.show()
    
    # Print correlation
    corr = df[[var, 'SalePrice']].corr().iloc[0,1]
    print(f'Correlation between {var} and SalePrice: {corr:.2f}\n')

# Categorical variables vs. SalePrice
cat_vars = ['Neighborhood', 'House Style']
for var in cat_vars:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=df[var], y=df['SalePrice'])
    plt.title(f'SalePrice by {var}')
    plt.xlabel(var)
    plt.ylabel('SalePrice')
    plt.xticks(rotation=45)
    plt.show()
    
    # Print mean SalePrice by category
    print(f'Mean SalePrice by {var}:')
    print(df.groupby(var)['SalePrice'].mean().sort_values(ascending=False).head())
    print() 
```

# Part 5: Correlation Heatmap and Outlier Detection

```python
# Ensure all inf values are replaced with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Correlation heatmap for numeric features (fully cleaned)
plt.figure(figsize=(14, 10))
corr = df.corr(numeric_only=True)
# Drop any rows or columns with NaN correlations
corr = corr.dropna(axis=0, how='any').dropna(axis=1, how='any')
sns.heatmap(corr, cmap='coolwarm', center=0, annot=False)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()

# Outlier detection for SalePrice and Gr Liv Area
outlier_vars = ['SalePrice', 'Gr Liv Area']
for var in outlier_vars:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[var].dropna(), orient='h')
    plt.title(f'Boxplot of {var} (Outlier Detection)')
    plt.xlabel(var)
    plt.show()
    
    # Show extreme values
    print(f'Top 5 highest values for {var}:')
    print(df[var].sort_values(ascending=False).head())
    print() 
```

# Part 6: Missing Value Analysis and Initial Feature Engineering

```python


# Summarize missing values (top 15 columns with most missing)
missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]
print('Columns with missing values (top 15):')
print(missing.head(15))

# Visualize missing data pattern (optional, requires missingno)
try:
    import missingno as msno
    msno.matrix(df)
    plt.title('Missing Data Matrix')
    plt.show()
    msno.heatmap(df)
    plt.title('Missing Data Heatmap')
    plt.show()
except ImportError:
    print('missingno not installed; skipping missing data visualization.')

# Suggest initial feature engineering ideas
print('Initial Feature Engineering Ideas:')
print('- Create TotalBathrooms = Full Bath + Half Bath*0.5 + Bsmt Full Bath + Bsmt Half Bath*0.5')
print('- Create Age = Yr Sold - Year Built')
print('- Encode categorical variables (e.g., Neighborhood, House Style) using one-hot encoding or label encoding')
print('- Consider combining low-frequency categories in categorical variables') 
```

## Part 7: Feature Engineering

```python



# Create TotalBathrooms feature
if all(col in df.columns for col in ['Full Bath', 'Half Bath', 'Bsmt Full Bath', 'Bsmt Half Bath']):
    df['TotalBathrooms'] = (
        df['Full Bath'] + 0.5 * df['Half Bath'] +
        df['Bsmt Full Bath'] + 0.5 * df['Bsmt Half Bath']
    )
    print('Created TotalBathrooms feature.')
else:
    print('Some bathroom columns missing; TotalBathrooms not created.')

# Create Age feature
if all(col in df.columns for col in ['Yr Sold', 'Year Built']):
    df['Age'] = df['Yr Sold'] - df['Year Built']
    print('Created Age feature.')
else:
    print('Yr Sold or Year Built column missing; Age not created.')

# Encode Neighborhood as an example (label encoding)
if 'Neighborhood' in df.columns:
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    df['Neighborhood_encoded'] = le.fit_transform(df['Neighborhood'].astype(str))
    print('Encoded Neighborhood as Neighborhood_encoded.')
else:
    print('Neighborhood column missing; encoding not performed.')

# Show updated DataFrame head
print('\nUpdated DataFrame with new features:')
display(df.head()) 
```

# Part 8: Hypothesis Formulation and Statistical Testing

```python



# Example hypotheses
print('Example Hypotheses:')
print('1. Houses in the "NridgHt" neighborhood have higher average SalePrice than those in "NAmes".')
print('2. Houses with more than 2 bathrooms have higher SalePrice than those with 2 or fewer.')
print('3. Newer houses (Age < 10 years) have higher SalePrice than older houses (Age >= 10 years).')

# Statistical test: t-test for SalePrice between two neighborhoods
from scipy.stats import ttest_ind

if all(col in df.columns for col in ['Neighborhood', 'SalePrice']):
    group1 = df[df['Neighborhood'] == 'NridgHt']['SalePrice'].dropna()
    group2 = df[df['Neighborhood'] == 'NAmes']['SalePrice'].dropna()
    if len(group1) > 0 and len(group2) > 0:
        t_stat, p_val = ttest_ind(group1, group2, equal_var=False)
        print(f'\nT-test: SalePrice in NridgHt vs. NAmes')
        print(f'T-statistic: {t_stat:.2f}, p-value: {p_val:.4f}')
        if p_val < 0.05:
            print('Result: Statistically significant difference in SalePrice between neighborhoods (p < 0.05).')
        else:
            print('Result: No statistically significant difference in SalePrice between neighborhoods (p >= 0.05).')
    else:
        print('Not enough data in one or both neighborhoods for t-test.')
else:
    print('Neighborhood or SalePrice column missing; t-test not performed.') 
```
